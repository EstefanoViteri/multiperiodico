```python
import random
random.seed(0)

```

```python
n_preventas = 10
min_recetas = 1
max_recetas = 5
min_tallos = 1
max_tallos = 24
min_bunches = 1
max_bunches = 24
min_boxes = 10
max_boxes = 100
min_flores = 1
max_flores = 5
flores = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M']
colores = ['α', 'β', 'γ', 'δ', 'ε', 'ζ', 'η', 'θ', 'ι', 'κ', 'λ', 'μ', 'ν']
grados = ['40','50','60']
variedades = ['Λ', 'Γ', 'Ξ', 'Π', 'Ρ', 'Σ', 'Τ', 'Υ', 'Φ', 'Χ', 'Ψ', 'Ω','Any']
tipos_de_caja = ['F','H','E','T']

dem = dict()
dem_info = dict()
sur_original ='Surtido 0'
for i in range(1,n_preventas+1):
    nom_prev ='Preventa_{}'.format(i)
    dem[nom_prev] = {sur_original: dict()}
    dem_info[nom_prev] = {'FechaUP': random.randint(1, 15),
                          'Boxes': random.randint(min_boxes, max_boxes),
                          'BoxCode': tipos_de_caja[random.randint(0, 3)],
                          #'Pack': ,
                          'Adelanto_max': random.randint(1, 2),
                          'Surtido_info': {sur_original: {'Receta_info': dict() } } }
    num_recetas = random.randint(min_recetas, max_recetas)
    for j in range(1,num_recetas+1):
        nom_rec = 'Receta_{}'.format(j)
        num_flores = random.randint(min_flores, max_flores)
        
        dem[nom_prev][sur_original][nom_rec] = {
            (a,flores[random.randint(0, 12)],
             colores[random.randint(0, 12)],
             grados[random.randint(0, 2)],
             variedades[random.randint(0, 12)]) 
            for a in range(1,num_flores+1)}
        
        dem_info[nom_prev]['Surtido_info'][sur_original]['Receta_info'][nom_rec] = {
            'Bunches': random.randint(min_bunches, max_bunches),
            'Flores_info': dict()
        }
        for k in dem[nom_prev][sur_original][nom_rec]:
            dem_info[nom_prev]['Surtido_info'][sur_original]['Receta_info'][nom_rec]['Flores_info'][k] = {
                'Tallos': random.randint(min_tallos,max_tallos)
            }
    
    
```
